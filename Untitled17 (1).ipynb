```python
print("The average age oh the patients:",df['age'].median())
```

```python
q1=df['age'].quantile(0.25)
q2=df['age'].quantile(0.75)
IQR=q2-q1
low=q1-1.5*IQR
up=q2+1.5*IQR
outliers=df[(df['age']<low)|(df['age']>up)]
print("Number of outliers:",len(outliers))
print("Outliers glucose values:\n",outliers['age'].values)
```

Since age has outliers we can take median value as the average.

```python
a=df[(df['age']>=25)&(df['age']<=29)]
print("Average glucose(Age 25-29):",a['glu'].mean())
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Average BMI:",df['bmi'].mean())
df['age_group']=pd.cut(df['age'],bins=[20,30,40,50,60,70,80],labels=['20-29','30-39','40-49','50-59','60-69','70-79'])
b=df.groupby('age_group')['bmi'].mean()
print("Average BMI by age group",b)

```

```python
q1=df['glu'].quantile(0.25)
q2=df['glu'].quantile(0.75)
IQR=q2-q1
low=q1-1.5*IQR
up=q2+1.5*IQR
outliers=df[(df['glu']<low)|(df['glu']>up)]
print("Number of outliers:",len(outliers))
print("Outliers glucose values:\n",outliers['glu'].values)
```

```python
print(df.std(numeric_only=True))
print("Highest variability",df.std(numeric_only=True).idxmax())
```

```python
print("Relation between glucose and bmi:",df['glu'].corr(df['bmi']))
```

Since the value is positive and moderate this has a high relation but not very strong

```python
print("Relation between age and pregnancy:",df['age'].corr(df['npreg']))
```

As the ralation is positive and near to one,there is strong relation between them and we can say that older women tend to have more pregnancy rates 

```python
print("Relation between insuline and glucose:",df['insulin'].corr(df['glu']))
```

The relation is positive but not very strong means higher glucose is somewhat linked to higher insuline but relation is noisy

```python
print("Relation between pregnancy and DiabetesPedigreeFunction:",df['npreg'].corr(df['ped']))
```

The relation is very close to zero so women with more pregnancies do not necessarily have higher diabetes pedigree function.

```python
y=df[df['age']<30]
o=df[df['age']>=30]
print("Average BMI(<30 years):",y['bmi'].mean())
print("Average BMI(>=30 years):",o['bmi'].mean())
```

So older people have slightly higher BMI

```python
print("Relation between blood pressure and glucose:",df['bp'].corr(df['glu']))
```

This relation is close to zero so it is not a strong relationship

```python
df['bmi_category']=pd.cut(df['bmi'],bins=[0,18.5,25,30,float('inf')],labels=['Underweight','Normal','Overweight','Obese'])
skin=df.groupby("bmi_category")['skin'].mean()
print("average skinthickness by bmi:",skin)
print("Normal skinthickness:",skin['Normal'])
print("obese skinthickness:",skin['Obese'])
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.boxplot(x=df['bmi_category'],y='skin',data=df,color='skyblue')
plt.title('Average skinthickness by BMI category')
plt.ylabel('Average skinthickness')
plt.xlabel('BMI category')
plt.show()
```

```python
plt.figure(figsize=(6,5))# Create a histogram
plt.hist(df['glu'],bins=5,color='pink', edgecolor='black')

# Add labels and title
plt.xlabel('Glucose')
plt.ylabel('Count')


# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='outcome',data=df,palette='Set2')
plt.title("Diabetes Outcome Count(0=No,1=Yes)")
plt.xlabel('Outcomes')
plt.ylabel('No of patients')
plt.show()
```

```python
sns.barplot(x='outcome',y='glu',data=df,palette='Set1')
plt.title("Average Glucose by Diabetes Outcome")
plt.xlabel('Outcomes(0=No,1=Yes)')
plt.ylabel('Average Glucose')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(),annot=True,cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```

Glucose has the strongest positive correltation with diabetes outcome.
BMI and Age also show some correlation with outcome.
Pregnancies has a weak but still positive relation.
