1.Target Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='target',data=df)
plt.title("Heart Disease count(0=No,1=Yes)")
plt.show()
```

Helps check class balance(how many with,without the heart disease)
More people are having the disease

2 Sex VS Heart Disease

```python
sns.countplot(x='sex',hue='target',data=df)
plt.title('Sex VS Heart Disease')
plt.show()
```

Males are having high number of heart disease while females have less rate

3.Age Distribution

```python
sns.histplot(df['age'],bins=10,kde=True)
plt.title('Age Distribution')
plt.show()
```

The people between Age group 50 and 60 are having the disease more

4 Age VS Disease

```python
sns.boxplot(x='target',y='age',data=df)
plt.title('Age VS Disease')
plt.show()
```

The people having 55 have High rate of disease

5 Cholesterol VS Disease

```python
sns.boxplot(x='target',y='chol',data=df)
plt.title('Cholesterol VS Disease')
plt.show()
```

```python
from scipy import stats
z_scores=stats.zscore(df['chol'])
abs_z_scores=abs(z_scores)
df_no_outliers=df[abs_z_scores<2]
```

Since it had some outliers,have to remove it

```python
sns.boxplot(x='target',y='chol',data=df_no_outliers)
plt.title('Cholesterol VS Disease')
plt.show()
```

The people having disease have an average Cholesterol rate of 240

6 Maximum Heart Rate VS Disease

```python
sns.boxplot(x='target',y='thalach ',data=df)
plt.title('Maximum Heart Rate VS Disease')
plt.show()
```

```python
z_scores=stats.zscore(df['thalach '])
abs_z_scores=abs(z_scores)
df_no_outliers=df[abs_z_scores<1.3]
```

```python
sns.boxplot(x='target',y='thalach ',data=df_no_outliers)
plt.title('Maximum Heart Rate VS Disease')
plt.show()
```

From the figure we can say that the people with the disease have the an average heart rate of 160 

7 Chest pain type VS Disease

```python
sns.countplot(x='cp',hue='target',data=df)
plt.title('Chest pain type VS Disease')
plt.show()
```

From this we can say the with type 2 cp have high rate of disease
People With type 0 have a low rate of disease

8 Correlation Heatmap

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(),annot=True,cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```

Chest pain type(cp) are highly linked with the heart disease.
Oldpeak(ST depression)-High values means high risk 
ca(number of major vessels)-High count means more risk 
Thalach(max heart rate)lower heart rate reserve,higher risk
Males are more Prone to disease than female
chol,resting blood pressure(resting_bp),fbs(fasting blood sugar) have a weak relation with disease
